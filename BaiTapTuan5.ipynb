[SumOfNumBers]
Viết chương trình Python sử dụng một trong 2 loại vòng lặp for hoặc while để
tính tống các số nguyên từ 1 đến n. (n <= 1000) 


```python
#Cách 1
n = int(input())
i = 1 
tong = 0
while i <= n :
    tong += i
    i+=1 
print(tong)
```

Kiểm tra số này có phải là số nguyên tố hay không? 

```python
while True : 
    n = int(input())
    if n > 0 : 
        break
if n < 2 : print('Khong phai so nguyen to')
else :
    for i in range(2 , int(n**0.5) +1) :
        if n%i == 0 : 
            print('Khong la so nguyen to.')
            break 
    else : print('la so nguyen to')
```

```python
#Cách 2 dùng hàm sẽ nhanh hơn
from math import *
def ham(n) :
    while True : 
        n = int(input())
        if n>0 : break 
    if n <2 : return print('Khong phai so nguyen to')
    for i in range(2, isqrt(n)+1 ) :
        if n%i == 0 : return f'khong la so nguyen to'
    return f'la so nguyen to'
n = int(input())
ham(n)
```

W4A6
"[SumPrimesInRange]
Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn
[a, b]. Mỗi dòng là một bộ test (a b)."

```python
from math import *
a, b = map(int,input().split())
tong = 0
for i in range(a,b+1) :
    if a < 2 : break
    for j in range(2 , isqrt(i)+1) :
        if i%j == 0 : break
    else : tong +=i 
print(tong)
```

```python
# Đây là bài toán nhập vào số nguyên n , tính tổng các chữ số nguyên tố trong n
from math import *
n = int(input())
tong = 0
while n!=0 :
    a = n%10 
    if a <2 : break
    for i in range(2 , isqrt(a) +1 ) :
        if a%i == 0 : break
    else : tong += a
    n//=10
print(tong)
```

[ClosePairNumber]
Hai số nguyên dương tạo thành một cặp số thân thiết khi chúng tuân theo quy luật:
Số này bằng tổng tất cả các ước của số kia (trừ chính nó) và ngược lại.
Ví dụ 220 và 284 là một cặp số thân thiết
220 ngoài chính nó ra, nó còn có 11 ước số khác là: 1, 2, 4, 5, 10, 11, 20, 44, 55,
110. Tổng của 11 ước này cho chúng ta 284
284 ngoài bản thân nó, còn có 5 ước số khác là: 1, 2, 4, 71, 142. Tổng của chúng
vừa đúng bằng 220
Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không.
Nếu có, in ra màn hình "true", ngược lại, in ra "false"

```python
a,b = map(int,input().split())
tong1 = 0 
tong2 = 0
for i in range(1,a) :
    if a%i == 0 : tong1+=i
for j in range(1,b) :
    if b%j == 0 : tong2+=j 
if a == tong2 and b == tong1 : 
    print('True')
else : print('False')
```
