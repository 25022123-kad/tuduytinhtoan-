# W3A17
Cấp số nhân
Cho 4 số a, b, c, d. Hãy kiểm tra xem 4 số này có thể theo thứ tự tạo thành 1 cấp số nhân với công bội nguyên theo đúng
thứ tự a, b, c, d hay không?
Gợi ý : Tìm công bội (b / a) nếu b chia hết cho a, sau đó lấy b nhân công bội và so sánh vs c, c nhân công bội và so sánh vs
d. 


```python
a, b, c, d = map(int,input().split())

if a!= 0 and b%a==0 : 
    q = b//a
    if c == b*q and d == c*q :
        print('Yes')
    else : print('No')
elif a==0 and b==0 and c==0 and d==0 : 
    print('yes')
else : print('No')
```

# W3A15
[Xếp loại học lực]

- >= 8.0: Giỏi
- >= 6.5: Khá
- >= 5.0: Trung bình
- < 5.0: Yếu 


```python
a, b, c, d = map(float,input().split())
TB = (a+b+c+d)/4
if TB >= 8 : print(f'{TB:.2f} Gioi')
elif 6.5<=TB<=8.0 : print(f'{TB:.2f} Kha')
elif 5<=TB<=6.5 : print(f'{TB:.2f} Trung Binh')
else  : print(f'{TB} Yeu')
```
