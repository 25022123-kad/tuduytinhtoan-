# Bài 7: Nhập vào số nguyên dương a, đếm số ước dương của a 


```python
a = int(input())
dem = 0
for i in range (1 , a+1) :
    if a % i == 0 :
        dem+= 1
        print (dem)
```

# Bài 5: Nhập vào số nguyên dương n. Tính S = 1 + 2 + 3 + 4 + … + n

```python
n =  int(input())
s = 0
i = 1
while i <= n :
    s += i 
    i += 1
    print (s)
```
